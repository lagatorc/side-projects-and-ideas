```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_percentage_error

import plotly.graph_objects as go
```

## Pytorch Data Processing

```python
class CementData(Dataset):
    def __init__(self, x, y):
        self.x = torch.tensor(x.values, dtype=torch.float)
        self.y = torch.tensor(np.expand_dims(y.values, 1), dtype=torch.float)
    def __len__(self):
        return len(self.x)
    def __getitem__(self,idx):
        features = self.x[idx,:]
        label = self.y[idx]
        return features, label
```

```python
train_data = CementData(x_train, y_train)
val_data = CementData(x_val, y_val)
test_data = CementData(x_test, y_test)

train_loader = DataLoader(train_data, batch_size=64)
val_loader = DataLoader(val_data, batch_size=len(val_data), shuffle=True)
test_loader = DataLoader(test_data, batch_size=len(test_data), shuffle=True)
```

## Model Training

```python
class ConcreteModel(nn.Module):
    def __init__(self):
        super(ConcreteModel, self).__init__()
        self.dense1 = nn.Linear(9, 36)
        self.dense2 = nn.Linear(36, 100)
        self.dense3 = nn.Linear(100, 36)
        self.dense4 = nn.Linear(36, 9)
        self.dense5 = nn.Linear(9, 1)
        
        self.dropout = nn.Dropout(p=0.1)
        
    def forward(self, x):
        x = F.tanh(self.dense1(x))
        x = self.dropout(x)
        x = F.tanh(self.dense2(x))
        x = self.dropout(x)
        x = F.tanh(self.dense3(x))
        x = self.dropout(x)
        x = F.tanh(self.dense4(x))
        x = self.dense5(x)
        return x
```

```python
# set up hyperopt hyperparameter tuning
```

```python
model = ConcreteModel()
optim = torch.optim.Adam(params = model.parameters(), lr = 0.003)
loss_func = nn.MSELoss()

train_losses = []
val_losses = []

EPOCH = 1000

for epoch in range(EPOCH):    
    batch_train_losses = []
    for train_feat, train_target in train_loader:
        train_pred = model(train_feat)
        train_loss = loss_func(train_pred, train_target)
        batch_train_losses.append(train_loss.detach().numpy())
        
        train_loss.backward()
        optim.step()
        optim.zero_grad()
        
    epoch_train_loss = sum(batch_train_losses) / len(batch_train_losses)
    train_losses.append(epoch_train_loss)
    
    batch_val_losses = []
    for val_feat, val_target in val_loader:
        val_pred = model(val_feat)
        val_loss = loss_func(val_pred, val_target)
        batch_val_losses.append(val_loss.detach().numpy())
    epoch_val_loss = sum(batch_val_losses) / len(batch_val_losses)
    val_losses.append(epoch_val_loss)
```

```python
for test_feat, test_lab in test_loader:
    test_lab = test_lab.detach().numpy()
    test_pred = model(test_feat).detach().numpy()
    r2 = r2_score(test_lab, test_pred)
    mape = mean_absolute_percentage_error(test_lab, test_pred)
    break
    
print(f"R2: {r2}")
print(f"MAPE: {mape}")
```

```python
fig = go.Figure(data=[go.Scatter(x = list(range(len(train_losses))), y=train_losses, name="Train"),
                     go.Scatter(x = list(range(len(val_losses))), y=val_losses, name="Validation")])
fig.show()
```
