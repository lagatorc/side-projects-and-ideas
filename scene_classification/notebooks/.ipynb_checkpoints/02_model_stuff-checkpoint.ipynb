```python
class Model(tf.keras.Model):
    def __init__(self):
        super(Model, self).__init__()
        self.scaling = tf.keras.layers.experimental.preprocessing.Rescaling(1./255, input_shape=(128, 128, 3))
        self.conv1 = tf.keras.layers.Conv2D(12, (3, 3), activation="relu")
        self.avg1 = tf.keras.layers.AveragePooling2D()
        self.conv2 = tf.keras.layers.Conv2D(24, (3, 3), activation="relu")
        self.avg2 = tf.keras.layers.MaxPooling2D()
        self.conv3 = tf.keras.layers.Conv2D(48, (3, 3), activation="relu")
        self.avg3 = tf.keras.layers.MaxPooling2D()
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = tf.keras.layers.Dense(64, activation="relu")
        self.dense2 = tf.keras.layers.Dense(32, activation="relu")
        self.dense3 = tf.keras.layers.Dense(5, activation="relu")
    def call(self, inputs):
        x = self.scaling(inputs)
        x = self.conv1(x)
        x = self.avg1(x)
        x = self.conv2(x)
        x = self.avg2(x)
        x = self.conv3(x)
        x = self.avg3(x)
        x = self.flatten(x)
        x = self.dense1(x)
        x = self.dense2(x)
        x = self.dense3(x)
        return x
```

```python
model = Model()
loss = tf.keras.losses.CategoricalCrossentropy(from_logits=True)
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', mode="min", min_delta=0)
opt = tf.keras.optimizers.RMSprop(learning_rate=0.0001)
```

```python
model.compile(optimizer=opt, loss=loss, metrics="acc")
```

```python
hist = model.fit(x=train_data,epochs=10, validation_data=val_data, verbose=2)
```

```python
plt.plot(hist.history["loss"])
plt.plot(hist.history["val_loss"])
```
